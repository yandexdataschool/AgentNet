```python
import pandas as pd
from scipy.io.arff import loadarff
#obtain pbcseq from http://www.openml.org/d/516
data,columns = loadarff("pbcseq.arff")
df = pd.DataFrame(data,columns=columns.names())
```

```python
del df["case_number"],df["number_of_days"],df["status"]

df = df.fillna(method='bfill')
df = df.replace('?',0)
df = df.replace('no',0)

df["age"] /= 365.

df["drug"] = df["drug"] == 'D-penicillamine'
df["sex"] = df["sex"]=="male"
df["presence_of_asictes"] = df["presence_of_asictes"]=='yes'
df["presence_of_hepatomegaly"] = df["presence_of_hepatomegaly"]=='yes'
df["presence_of_spiders"] = df["presence_of_spiders"]==1


df["day"] = df["day"].astype('int')

df = df.astype('float32')
```

```python
df.to_csv("pbcseq.csv",index=None)
```
